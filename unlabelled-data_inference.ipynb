```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
raw_data: pd.DataFrame = pd.read_csv(
    filepath_or_buffer="../../../data/MOS2E03230475_30Hz_proc.csv",
    index_col=0,
    nrows=100,
    # nrows=4559526,
)
```

```python
raw_data.info()
```

```python
raw_data.sort_values(
    by=["Timestamp"],
    ascending=True,
    inplace=True,
)
```

```python
raw_data["Date"] = pd.to_datetime(
    raw_data["Timestamp"],
).dt.strftime(date_format="%d/%m/%Y")

raw_data["Time"] = pd.to_datetime(
    raw_data["Timestamp"],
).dt.strftime(date_format="%H:%M:%S")
```

```python
from itertools import cycle

c: cycle = cycle(range(1, 31))
raw_data["ID"] = [next(c) for _ in range(len(raw_data))]
```

```python
raw_data.drop(
    columns=[
        "SensorID",
        "HalterID",
        "CollarID",
        "CowID",
        "Orientation",
        "Timestamp",
    ],
    inplace=True,
)

raw_data.rename(
    columns={
        "X": "X_axis",
        "Y": "y_axis",
        "Z": "z_axis",
    },
    inplace=True,
)
```

```python
column_order: list[str] = ["Time", "ID", "Date", "X_axis", "y_axis", "z_axis"]
raw_data = raw_data.reindex(columns=column_order)
```
